```python
import pandas as pd
import numpy as np
```

## Position

```python
df = pd.read_csv('data/FullData.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['National_Kit', 'Weight', 'Height', 'Weak_foot', 'Club_Kit', 'Skill_Moves', 'Club_Joining', 'Contract_Expiry', 'Birth_Date', 'Age', 'Work_Rate', 'Preffered_Position'], axis=1)
df[df['Club_Position'].isnull()]
```

```python
df.dropna(subset=['Club_Position'], how='all', inplace=True) # Remove player above (FYI he is ST)
df[df['National_Position'].notnull()]
```

Since National_Positions are empty for most rows, We know that almost all players have same position for Club and National Matches.
We will remove National_Positions and keep the target as Club_Positions.

```python
df = df.drop(['National_Position'], axis=1)
df.shape
```

```python
unique_CP = df['Club_Position'].unique()
unique_CP
```

```python
# Grouping positions
attack = ['LW', 'RW', 'ST', 'RS', 'LF', 'CF', 'RF', 'LS']
mid = ['RCM', 'CAM', 'LCM', 'LM', 'LDM', 'RM', 'CDM', 'RDM', 'LAM', 'CM', 'RAM']
defend = ['LCB', 'RB', 'RCB', 'LB', 'CDM', 'RDM', 'CB', 'RWB', 'LWB']
gk = ['GK']
others = ['Sub', 'Res'] # Remove these rows
```

```python
df = df[(df['Club_Position'] != others[0]) & (df['Club_Position'] != others[1])]
df.shape
```

```python
df['Preffered_Foot_Num'] = None # 0 -> Right, 1 -> Left
def categorize_preffered_foot(row):
    if row['Preffered_Foot'] == 'Right':
        return 0
    else:
        return 1
df['Preffered_Foot_Num'] = df.apply(categorize_preffered_foot, axis=1)
```

## Simple Predictions -> 4 labels only

```python
df1 = df.drop(['Preffered_Foot', 'Preffered_Foot_Num'], axis=1)
```

```python
df1['Club_Position_Num'] = None # 0 -> Keeper, 1 -> Defenders, 2 -> Mid, 3-> Attackers
def categorize_club_positions(row):
    if row['Club_Position'] in attack:
        return 3
    elif row['Club_Position'] in mid:
        return 2
    elif row['Club_Position'] in defend:
        return 1
    else:
        return 0
df1['Club_Position_Num'] = df1.apply(categorize_club_positions, axis=1)
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test1 = train_test_split(df1, test_size=0.2)
```

```python
X_train.shape, X_test1.shape
```

```python
y_train = X_train['Club_Position_Num']
X_train = X_train.drop(['Nationality', 'Club', 'Club_Position_Num', 'Club_Position', 'Name'], axis=1)
```

```python
y_test = X_test1[['Club_Position_Num']]
```

```python
X_test = X_test1.drop(['Nationality', 'Club', 'Club_Position_Num', 'Club_Position', 'Name'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression().fit(X_train, y_train)
linear_regression_predictions = linear_regression.predict(X_test)
linear_regression_score = linear_regression.score(X_test, y_test)
linear_regression_score_train = linear_regression.score(X_train, y_train)
linear_regression_score, linear_regression_score_train
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression().fit(X_train, y_train)
logistic_regression_predictions = logistic_regression.predict(X_test)
logistic_regression_score = logistic_regression.score(X_test, y_test)
logistic_regression_score_train = logistic_regression.score(X_train, y_train)
logistic_regression_score, logistic_regression_score_train
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_classifier = DecisionTreeClassifier().fit(X_train, y_train)
decision_tree_predictions = decision_tree_classifier.predict(X_test)
decision_tree_score = decision_tree_classifier.score(X_test, y_test)
decision_tree_score_train = decision_tree_classifier.score(X_train, y_train)
decision_tree_score, decision_tree_score_train
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier().fit(X_train, y_train)
random_forest_predictions = random_forest_classifier.predict(X_test)
random_forest_score = random_forest_classifier.score(X_test, y_test)
random_forest_score_train = random_forest_classifier.score(X_train, y_train)
random_forest_score, random_forest_score_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
kneighbors_classifier = KNeighborsClassifier().fit(X_train, y_train)
kneighbors_predictions = kneighbors_classifier.predict(X_test)
kneighbors_score = kneighbors_classifier.score(X_test, y_test)
kneighbors_score_train = kneighbors_classifier.score(X_train, y_train)
kneighbors_score, kneighbors_score_train
```

```python
from sklearn.svm import SVC
sv_classifier = SVC().fit(X_train, y_train)
svc_predictions = sv_classifier.predict(X_test)
svc_score = sv_classifier.score(X_test, y_test)
svc_score_train = sv_classifier.score(X_train, y_train)
svc_score, svc_score_train
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB

gaussian_nb_classifier = GaussianNB().fit(X_train, y_train)
gaussian_nb_predictions = gaussian_nb_classifier.predict(X_test)
gaussian_nb_score = gaussian_nb_classifier.score(X_test, y_test)
gaussian_nb_score_train = gaussian_nb_classifier.score(X_train, y_train)

multinomial_nb_classifier = MultinomialNB().fit(X_train, y_train)
multinomial_nb_predictions = multinomial_nb_classifier.predict(X_test)
multinomial_nb_score = multinomial_nb_classifier.score(X_test, y_test)
multinomial_nb_score_train = multinomial_nb_classifier.score(X_train, y_train)

(gaussian_nb_score, gaussian_nb_score_train), (multinomial_nb_score, multinomial_nb_score_train)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100,10)).fit(X_train, y_train)
mlp_predictions = mlp_classifier.predict(X_test)
mlp_score = mlp_classifier.score(X_test, y_test)
mlp_score_train = mlp_classifier.score(X_train, y_train)
mlp_score, mlp_score_train
```

```python
X_test1['MLP_Prediction'] = mlp_predictions
def rev_categorize_club_positions(row):
    if row['MLP_Prediction'] == 3:
        return 'ATT'
    elif row['MLP_Prediction'] == 2:
        return 'MID'
    elif row['MLP_Prediction'] == 1:
        return 'DEF'
    else:
        return 'GK'
X_test1['MLP_Prediction_Rev'] = X_test1.apply(rev_categorize_club_positions, axis=1)
```
